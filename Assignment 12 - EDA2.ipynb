```python
# Apply scaling techniques
numeric_features = ['age' , 'fnlwgt' , 'education_num' , 'capital_gain' , 'capital_loss' , 'hours_per_week']
```

```python
# Standard Scaling
standard_scaler = StandardScaler()
data_standard_scaled = data.copy()
data_standard_scaled[numeric_features] = standard_scaler.fit_transform(data[numeric_features])
```

```python
# Min-Max Scaling
min_max_scaler = MinMaxScaler()
data_min_max_scaled = data.copy()
data_min_max_scaled[numeric_features] = min_max_scaler.fit_transform(data[numeric_features])
```

# 2. Encoding Techniques

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
# Identify categorical features
categorical_features = data.select_dtypes(include=['object']).columns.tolist()
categorical_features.remove('income')  # Remove target variable if it is categorical
```

```python
# Apply One-Hot Encoding to categorical variables with less than 5 categories
one_hot_encoder = OneHotEncoder(drop='first', sparse=False)
one_hot_features = ['sex', 'race', 'relationship']
```

```python
# Apply Label Encoding to categorical variables with more than 5 categories
label_encoder = LabelEncoder()
label_encoded_features = ['workclass', 'education', 'marital_status', 'occupation', 'native_country']
```

```python
# Create copies of the dataset for encoding
data_one_hot_encoded = data.copy()
data_label_encoded = data.copy()
```

```python
# One-Hot Encoding
data_one_hot_encoded = pd.get_dummies(data_one_hot_encoded, columns=one_hot_features, drop_first=True)
```

```python
# Label Encoding
for feature in label_encoded_features:
    data_label_encoded[feature] = label_encoder.fit_transform(data_label_encoded[feature])
```

```python
# Display the encoded data
data_one_hot_encoded.head()
data_label_encoded.head()
```

# 3. Feature Engineering

```python
import numpy as np
```

```python
# Create new features
data_engineered = data.copy()
```

```python
# Feature 1: Age Group (binned age into categories)
bins = [0, 18, 30, 40, 50, 60, np.inf]
labels = ['0-17', '18-29', '30-39', '40-49', '50-59', '60+']
data_engineered['age_group'] = pd.cut(data_engineered['age'], bins=bins, labels=labels, right=False)
```

```python
# Feature 2: Capital Gain to Loss Ratio
data_engineered['capital_gain_loss_ratio'] = data_engineered['capital_gain'] / (data_engineered['capital_loss'] + 1)  # Added 1 to avoid division by zero
```

```python
# Apply a transformation to a skewed numerical feature
# Checking skewness
skewness = data_engineered['capital_gain'].skew()
print(f"Skewness of 'capital_gain': {skewness}")
```

```python
# Apply log transformation
data_engineered['capital_gain_log'] = np.log1p(data_engineered['capital_gain'])  # log1p is used to handle zero values
```

# 4. Feature Selection:

```python
from sklearn.ensemble import IsolationForest
```

```python
# Using Isolation Forest to identify outliers
iso_forest = IsolationForest(contamination=0.05, random_state=42)  # Adjust contamination based on dataset
data_engineered['outlier'] = iso_forest.fit_predict(data_engineered[numeric_features])
```

```python
# Removing outliers
data_no_outliers = data_engineered[data_engineered['outlier'] == 1].drop(columns=['outlier'])
```

Discussion on Outliers:

Outliers can significantly affect model performance by:

Biasing the Model: Outliers can skew the distribution of the data, leading to biased model parameters.

Reducing Model Accuracy: The presence of outliers can lead to overfitting, where the model performs well on training data but poorly on unseen data.

Impact on Metrics: Metrics such as mean squared error (MSE) and R-squared can be heavily influenced by outliers, leading to misleading evaluations of model performance.
