# Task 2 : Dimensionality Reduction with PCA

```python
#standardize the features
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Separating out the features
features = ['Alcohol', 'Malic', 'Ash', 'Alcalinity', 'Magnesium', 'Phenols', 
            'Flavanoids', 'Nonflavanoids', 'Proanthocyanins', 'Color', 'Hue', 
            'Dilution', 'Proline']
x = wine_data.loc[:, features].values

# Standardize the features
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# PCA with all components
pca = PCA()
pca.fit(x_scaled)

# Scree plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(features)+1), pca.explained_variance_ratio_, marker='o', linestyle='--')
plt.title('Scree Plot')
plt.xlabel('Number of Principal Components')
plt.ylabel('Variance Explained')
plt.show()

# Cumulative explained variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(features)+1), cumulative_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Variance Explained')
plt.axhline(y=0.95, color='r', linestyle='--')  # threshold for 95% variance
plt.show()
```

```python
# Applying PCA with the chosen number of components
optimal_components = 2  # Adjust based on the scree plot/cumulative variance analysis
pca = PCA(n_components=optimal_components)
principal_components = pca.fit_transform(x_scaled)

# Creating a DataFrame with the principal components
pca_df = pd.DataFrame(data=principal_components, columns=[f'Principal Component {i+1}' for i in range(optimal_components)])

# Concatenating the Type column with the principal components DataFrame for visualization
final_df = pd.concat([wine_data[['Type']], pca_df], axis=1)
print(final_df.head())
```

```python
# Visualizing the principal components
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Principal Component 1', y='Principal Component 2', hue='Type', data=final_df, palette='viridis')
plt.title('PCA of Wine Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

# Task 3: Clustering with Original Data

```python
#Perform clustering on the original wine dataset and evaluate its performance
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
# Applying K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # Assuming 3 clusters, adjust if necessary
clusters = kmeans.fit_predict(x_scaled)

# Adding the cluster labels to the original dataframe
wine_data['Cluster'] = clusters
```

```python
# Visualizing the clusters in a 2D plot using PCA for dimensionality reduction
pca = PCA(n_components=2)
pca_components = pca.fit_transform(x_scaled)
pca_df = pd.DataFrame(data=pca_components, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['Cluster'] = clusters

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Principal Component 1', y='Principal Component 2', hue='Cluster', data=pca_df, palette='viridis')
plt.title('K-means Clustering of Wine Dataset (PCA-reduced)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# Silhouette score
silhouette_avg = silhouette_score(x_scaled, clusters)
print(f'Silhouette Score: {silhouette_avg:.2f}')

# Davies-Bouldin index
davies_bouldin = davies_bouldin_score(x_scaled, clusters)
print(f'Davies-Bouldin Index: {davies_bouldin:.2f}')
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
# Hierarchical Clustering on Original Data
linked = linkage(x_scaled, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrogram for Original Wine Dataset')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

# Task 4: Clustering with PCA Data

```python
# Applying K-means clustering to PCA-transformed data
kmeans_pca = KMeans(n_clusters=3, random_state=42)  # Assuming 3 clusters, adjust if necessary
clusters_pca = kmeans_pca.fit_predict(pca_df)

# Adding the cluster labels to the PCA DataFrame
pca_df['Cluster'] = clusters_pca
```

```python
# Evaluate clustering performance on PCA-transformed data
silhouette_avg_pca = silhouette_score(pca_df[['Principal Component 1', 'Principal Component 2']], clusters_pca)
davies_bouldin_pca = davies_bouldin_score(pca_df[['Principal Component 1', 'Principal Component 2']], clusters_pca)

print(f'Silhouette Score (PCA): {silhouette_avg_pca:.2f}')
print(f'Davies-Bouldin Index (PCA): {davies_bouldin_pca:.2f}')
```

# Task 5: Comparison and Analysis

1. Compare the Clustering Results
The clustering results on the original dataset and the PCA-transformed dataset are as follows:

Original Dataset:

Silhouette Score: 0.28
Davies-Bouldin Index: 1.39
PCA-Transformed Dataset:

Silhouette Score: 0.56
Davies-Bouldin Index: 0.60

2. Similarities and Differences in Clustering Results
Similarities:

Both clustering analyses used the same algorithm (K-means) and the same number of clusters (3).
The general approach of clustering remains consistent, involving the grouping of similar data points.
Differences:

Silhouette Score: The PCA-transformed data resulted in a significantly higher silhouette score (0.56) compared to the original dataset (0.28). A higher silhouette score indicates that the clusters are more well-defined and separated.
Davies-Bouldin Index: The PCA-transformed data resulted in a lower Davies-Bouldin Index (0.60) compared to the original dataset (1.39). A lower Davies-Bouldin Index suggests better clustering performance, with less intra-cluster variance and better separation between clusters.

3. Impact of Dimensionality Reduction on Clustering Performance
The application of PCA before clustering had a noticeable positive impact on clustering performance:

Improved Cluster Separation: The PCA-transformed dataset resulted in clusters that are more distinct and well-separated, as indicated by the higher silhouette score and lower Davies-Bouldin Index.
Noise Reduction: PCA helps to remove noise and irrelevant features, which can lead to more accurate and meaningful clusters. By focusing on the principal components that capture the most variance, PCA simplifies the dataset and enhances clustering performance.
Computational Efficiency: Reducing the number of dimensions can also make the clustering algorithm more efficient, as it has to process fewer features.

4. Trade-offs Between Using PCA and Clustering Directly on the Original Dataset

Advantages of Using PCA

Improved Performance: As observed, PCA can enhance clustering performance by focusing on the most important features.
Noise Reduction: PCA helps in reducing noise and irrelevant features, leading to more meaningful clusters.
Visualization: PCA reduces the dimensionality, making it easier to visualize high-dimensional data in 2D or 3D.

Disadvantages of Using PCA

Loss of Information: PCA involves dimensionality reduction, which can lead to the loss of some information. While PCA tries to retain the most important variance, some minor but potentially important details might be lost.
Interpretability: Principal components are linear combinations of the original features, making it harder to interpret the results in terms of the original features.
Parameter Selection: Choosing the number of principal components can be subjective and may require trial and error.

Advantages of Clustering on Original Dataset

Complete Information: All original features are considered, potentially capturing all aspects of the data.
Interpretability: Clusters can be directly interpreted in terms of the original features.

Disadvantages of Clustering on Original Dataset

Noise and Redundancy: High-dimensional data may contain noise and redundant features that can degrade clustering performance.
Computational Complexity: High-dimensional data increases the computational complexity of clustering algorithms.

2. Practical Implications of Using PCA and Clustering in Data Analysis

Improved Data Quality: PCA can enhance the quality of data by eliminating noise and redundant features, leading to more accurate clustering results.

Enhanced Interpretability: For high-dimensional data, PCA simplifies the dataset, making it easier to understand and visualize patterns and clusters.

Efficiency: PCA reduces the computational complexity of clustering algorithms by lowering the number of dimensions, which can be particularly beneficial for large datasets.

Feature Selection: PCA can serve as a tool for feature selection, identifying the most important features that contribute to the variance in the data.

3. Recommendations for Using Each Technique

When to Use PCA:

High-Dimensional Data: When dealing with datasets with a large number of features, PCA can help reduce dimensionality and improve clustering performance.

Noise Reduction: If the dataset contains noise or irrelevant features, PCA can help eliminate these and focus on the most significant features.

Visualization Needs: When there is a need to visualize high-dimensional data, PCA can reduce the dimensions to 2 or 3, facilitating easier interpretation and visualization.

Preprocessing Step: PCA can be used as a preprocessing step before applying clustering or other machine learning algorithms to enhance their performance.

When to Cluster Directly on Original Data:

Low-Dimensional Data: If the dataset has a manageable number of features and dimensionality reduction is not necessary.

Full Information Retention: When retaining all original information is critical and there is a risk of losing important 
details through PCA.

Interpretability: When it is essential to interpret the clusters in terms of the original features for better understanding and communication of results.

Simplicity: For straightforward clustering tasks where the data is already well-structured and clean, clustering directly on the original data may be simpler and sufficient.
