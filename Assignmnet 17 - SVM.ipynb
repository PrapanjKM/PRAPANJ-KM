```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

```python
# Define the Cramér's V function
def cramers_v(confusion_matrix):
    # Ensure the confusion matrix is not empty
    if confusion_matrix.size == 0:
        return 0
    chi2, _, _, _ = ss.chi2_contingency(confusion_matrix)
    n = confusion_matrix.sum().sum()
    k = min(confusion_matrix.shape) - 1
    return np.sqrt(chi2 / (n * k))
```

```python
columns = df.columns
```

```python
# Initialize the correlation matrix
corr_matrix = pd.DataFrame(index=columns, columns=columns, dtype=float)
```

```python
# Compute Cramér's V for each pair of features
for col1 in columns:
    for col2 in columns:
        confusion_matrix = pd.crosstab(df[col1], df[col2])
        value = cramers_v(confusion_matrix)
        corr_matrix.loc[col1, col2] = value
```

```python
# Plot the correlation heatmap
plt.figure(figsize=(15, 15))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=0, vmax=1)
plt.title('Cramér\'s V Correlation Matrix')
plt.show()
```

# Task 4: SVM Implementation

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Separate features and target variable
X = df.drop(columns='class')  # Replace 'class' with the actual target column name
y = df['class']
```

```python

# One-hot encode categorical features
encoder = OneHotEncoder(sparse=False, drop='first')
X_encoded = encoder.fit_transform(X)
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.3, random_state=42)
```

```python
# Initialize and train the SVM classifier
svm_model = SVC(kernel='linear', random_state=42)  # You can choose other kernels like 'rbf'
svm_model.fit(X_train, y_train)
```

```python
# Predict on the testing data
y_pred = svm_model.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:")
print(report)
```

# Task 5: Visualization of SVM Results


```python
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
# Encode categorical features
encoder = OneHotEncoder(sparse=False, drop='first')
X_encoded = encoder.fit_transform(X)
```

```python
# Encode target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
# Standardize features before PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)
```

```python
# Apply PCA to reduce data to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Split the reduced data
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y_encoded, test_size=0.3, random_state=42)
```

```python
# Train SVM on PCA-reduced data
svm_model_pca = SVC(kernel='linear', random_state=42)
svm_model_pca.fit(X_train_pca, y_train)
```

```python
# Plot decision boundaries
plt.figure(figsize=(10, 8))
```

```python
# Define the colors
cmap = ListedColormap(['#FFAAAA', '#AAFFAA'])
cmap_background = ListedColormap(['#FFAAAA', '#AAFFAA'])
```

```python
# Create a mesh grid
x_min, x_max = X_test_pca[:, 0].min() - 1, X_test_pca[:, 0].max() + 1
y_min, y_max = X_test_pca[:, 1].min() - 1, X_test_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))
```

```python
# Predict using the trained model
Z = svm_model_pca.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
# Plot decision boundaries
plt.contourf(xx, yy, Z, alpha=0.3, cmap=cmap_background)
```

```python
# Plot the test points
plt.scatter(X_test_pca[:, 0], X_test_pca[:, 1], c=y_test, edgecolor='k', s=50, cmap=cmap)
plt.title('SVM Decision Boundary with PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

# Task 6: Parameter Tuning and Optimization

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
# Define the parameter grid
param_grid = {
    'svm__kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'svm__C': [0.1, 1, 10, 100],
    'svm__gamma': ['scale', 'auto']  # Used for 'rbf', 'poly', 'sigmoid'
}
```

```python
# Create a pipeline with scaling and SVM
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', SVC())
])
```

```python
# Initialize GridSearchCV
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
```

```python
# Fit GridSearchCV
grid_search.fit(X_train, y_train)
```

```python
# Best parameters and best score
print(f"Best parameters: {grid_search.best_params_}")
print(f"Best cross-validation score: {grid_search.best_score_:.2f}")
```

```python
# Evaluate on the test set
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
print(f"Test set accuracy: {accuracy_score(y_test, y_pred):.2f}")
```

3. Discuss Practical Implications of SVM in Real-World Classification Tasks

Answer:

Support Vector Machines (SVMs) offer powerful classification capabilities, especially for high-dimensional and complex datasets. They excel in scenarios with clear margins of separation, making them ideal for tasks like text classification and image recognition where data is inherently high-dimensional. SVMs are effective with smaller to medium-sized datasets but can become computationally intensive and memory-heavy with very large datasets, necessitating careful consideration of computational resources. They require thorough feature engineering and preprocessing to handle unstructured data effectively. While SVMs can be adapted for multi-class classification through strategies like one-vs-one or one-vs-all, their practical application often involves careful tuning of hyperparameters, including kernel type and regularization parameters. Overall, SVMs are best suited for problems where the data dimensionality and complexity justify their use, but for extremely large datasets or real-time applications, alternative models may be more efficient.
